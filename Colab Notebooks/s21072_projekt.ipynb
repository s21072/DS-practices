```python
import os
from google.colab import drive
drive.mount('/content/drive')
os.chdir('/content/drive/My Drive/SUML_projekt')
path = Path('/content/drive/My Drive/SUML_projekt')
```

```python
classes = ['American Bobtail', 'American Shorthair', 'Birman', 'Chartreux', 'Domestic shorthair']
```

```python
path = Path('treningowe') # podajemy nazwę folderu, w którym znajdą się pliki
if not path.exists(): # folder zostanie dla nas stworzony, jeżeli jeszcze go nie mamy
  path.mkdir()
  for o in classes:
    dest = (path/o)
    dest.mkdir(exist_ok=True)
    urls = search_images_ddg(f' {o} cat', max_images=100) # w miejsce keyword podajemy dodatkowe słowo kluczowe do wyszukiwarki np. "zwierzaki"
    download_images(dest, urls=urls)
```

```python
data = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2,seed=42), #valid_pct - 20% danych do testowania modelu
    get_y=parent_label,
    item_tfms=Resize(512)
)
```

```python
data = data.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5), # Określamy ilość materiałów treningowych oraz walidacyjnych, wielkość zdjęć, augmentację
    batch_tfms=aug_transforms(max_rotate=2.0)
)
dls = data.dataloaders(path, bs = 32, num_workers = 0) #batch domyślny
```

```python
learn = vision_learner(dls, alexnet, metrics=accuracy)
```

```python
learn.fit(5)
```

```python
learn.save('stage-1')
```

```python
learn.export(fname='model_ropoznawanie_kotow.pkl') #źle działa
```
