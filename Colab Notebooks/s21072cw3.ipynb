```python
#The mean age and standard deviation in two separate gorups (men&women)
print(round(df_8['Age'].groupby(df_8['Sex']).mean(),2))
print(round(df_8['Age'].groupby(df_8['Sex']).std(),2)) 
```

```python
#the percentage of men in the dataset
counter = 0
for pp in df_8['Sex']:
  if (pp == 'M'):
    counter+=1
print(round(counter/len(df_8.index)*100,2), '%')
```

```python
#the number of women aged 45 to 50 years
counter = 0
for pp in df_8[df_8['Sex']=='F']['Age']:
  if (45 < pp < 50):
    counter+=1
print(counter)
```
