#### 2.1. Sieƒá nieukierunkowana (symetryczna)
##### üá¨üáß Symmetric network

```python
x_symetryczna = nx.Graph()
# w tym miejscu tworzymy nowƒÖ zmiennƒÖ, kt√≥ra bƒôdzie korzysta≈Ça z networkx 

x_symetryczna.add_edge('Kyran','Ahsis')
x_symetryczna.add_edge('Furrows','Kyran')
x_symetryczna.add_edge('Palaved','Aloy')
x_symetryczna.add_edge('Aloy','Kyran')
x_symetryczna.add_edge('Kyran','Palaved')
x_symetryczna.add_edge('Kyran','Tufanah')
x_symetryczna.add_edge('Erend','Sobeck')
x_symetryczna.add_edge('Sobeck','Palaved')
x_symetryczna.add_edge('Kyran','Sobeck')
x_symetryczna.add_edge('Palaved','Tufanah')
x_symetryczna.add_edge('Aratak','Tufanah')
x_symetryczna.add_edge('Avad','Tufanah')
x_symetryczna.add_edge('Kyran','Erend')
x_symetryczna.add_edge('Avad','Kyran')
# za pomocƒÖ .add_edge dodajemy do naszej sieci kolejne relacje oraz nody
```

```python
nx.draw_networkx(x_symetryczna)
# sprawdzamy jak wyglƒÖda nasza sieƒá
```

```python
nx.spring_layout(x_symetryczna)
# mo≈ºemy podejrzeƒá¬†jak zapisana jest nasza sieƒá¬†spo≈Çeczna
```

```python
nx.info(x_symetryczna)
```

#### 2.2. Sieƒá ukierunkowana (asymetryczna)
##### üá¨üáß Asymmetric network

```python
y_asymetryczna = nx.DiGraph()
# zrobimy co≈õ podobnego, ale tym razem stworzona przez nas sieƒá bƒôdzie ukierunkowana

y_asymetryczna.add_edge('Kyran','Ahsis')
y_asymetryczna.add_edge('Furrows','Kyran')
y_asymetryczna.add_edge('Palaved','Aloy')
y_asymetryczna.add_edge('Aloy','Kyran')
y_asymetryczna.add_edge('Kyran','Palaved')
y_asymetryczna.add_edge('Kyran','Tufanah')
y_asymetryczna.add_edge('Erend','Sobeck')
y_asymetryczna.add_edge('Sobeck','Palaved')
y_asymetryczna.add_edge('Kyran','Sobeck')
y_asymetryczna.add_edge('Palaved','Tufanah')
y_asymetryczna.add_edge('Aratak','Tufanah')
y_asymetryczna.add_edge('Avad','Tufanah')
y_asymetryczna.add_edge('Kyran','Erend')
y_asymetryczna.add_edge('Avad','Kyran')
```

```python
nx.draw_networkx(y_asymetryczna)

# nx.draw_networkx(y_asymetryczna, arrowsize = 30, with_labels = False, node_size = 500, node_color = "#1fb45b", edge_color = "#b41f2e")
```

```python
nx.spring_layout(y_asymetryczna)
```

#### 2.3. Sieƒá wa≈ºona
##### üá¨üáß Weighted network

```python
z_wazona = nx.Graph()

z_wazona.add_edge('Kyran','Ahsis', weight = 1)
z_wazona.add_edge('Furrows','Kyran')
z_wazona.add_edge('Palaved','Aloy')
z_wazona.add_edge('Aloy','Kyran')
z_wazona.add_edge('Kyran','Palaved')
z_wazona.add_edge('Kyran','Tufanah')
z_wazona.add_edge('Erend','Sobeck')
z_wazona.add_edge('Sobeck','Palaved')
z_wazona.add_edge('Kyran','Sobeck')
z_wazona.add_edge('Palaved','Tufanah')
z_wazona.add_edge('Aratak','Tufanah')
z_wazona.add_edge('Avad','Tufanah')
z_wazona.add_edge('Kyran','Erend')
z_wazona.add_edge('Avad','Kyran')
```

```python
nx.draw_networkx(z_wazona)
```

```python
nx.spring_layout(z_wazona)
```

### 3. Podstawowe metryki sieci
#### üá¨üáß Basic metrics
Teraz przejdziemy do analizy podstawowych metryk opisujƒÖcych nasze trzy¬†sieci (ukierunkowanƒÖ, nieukierunkowanƒÖ i wa≈ºonƒÖ).

```python
print(nx.degree(x_symetryczna, "Tufanah"))
print(nx.degree(y_asymetryczna, "Aloy"))
print(nx.degree(z_wazona, "Aloy"))
# warto≈õƒá¬†ta jest taka sama w przypadku wszystkich sieci
# nodal degree znany z czƒô≈õci teoretycznej
```

```python
print(nx.average_node_connectivity(x_symetryczna))
print(nx.average_node_connectivity(y_asymetryczna))
print(nx.average_node_connectivity(z_wazona))
```

```python
print(nx.average_shortest_path_length(x_symetryczna))
print(nx.average_shortest_path_length(y_asymetryczna))
print(nx.average_shortest_path_length(z_wazona))
```

```python
print(nx.clustering(x_symetryczna,"Ahsis"))
print(nx.clustering(y_asymetryczna,"Ahsis"))
print(nx.clustering(z_wazona,"Ahsis"))

# warto≈õƒá¬†ta jest taka sama w przypadku wszystkich sieci
# w sieciach spo≈Çecznych wystƒôpuje tendencja do tworzenia klastr√≥w (czyli "podgrup")
# mamy dwie mo≈ºliwo≈õci albo liczymy average_clustering dla ca≈Çej sieci lub dla konkretnej osoby
# co to oznacza? odnosi siƒô do frakcji relacji dla noda (osoby, kt√≥rƒÖ analizujemy) oraz jego znajomych (po≈ÇƒÖcze≈Ñ)
```

```python
print(nx.average_clustering(x_symetryczna))
print(nx.average_clustering(y_asymetryczna))
print(nx.average_clustering(z_wazona))

# warto≈õƒá¬†ta jest taka sama w przypadku wszystkich sieci
# warto≈õci ze wszystkich nod√≥w sƒÖ sumowane i dzielone przez liczbƒô¬†wƒôz≈Ç√≥w w sieci
```

```python
print(nx.shortest_path(x_symetryczna, "Ahsis", "Palaved"))
print(nx.shortest_path(y_asymetryczna, "Aloy", "Tufanah"))
print(nx.shortest_path(z_wazona, "Ahsis", "Palaved"))

# warto≈õƒá¬†ta jest taka sama w przypadku wszystkich sieci
# szukamy najkr√≥tszej ≈õcie≈ºki miƒôdzy poszczeg√≥lnymi elementami z naszej sieci
```

```python
print(nx.eccentricity(x_symetryczna, 'Aloy'))
print(nx.eccentricity(z_wazona, 'Aloy'))

# warto≈õƒá¬†ta jest taka sama w przypadku wszystkich sieci
# eccentricity odnosi siƒô¬†do najwiƒôkszego dystansu miƒôdzy nodem, dla kt√≥rego liczymy i wszystkimi innymi nodami
```

```python
dcolor = dict(x.degree)
low, *_, high = sorted(dcolor.values())
norm = mpl.colors.Normalize(vmin=low, vmax=high, clip=True)
mapper = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.coolwarm)
```

```python
pos = nx.spring_layout(x)
node_color = [mapper.to_rgba(i) for i in dcolor.values()]
node_size =  35
plt.figure(figsize=(15,15))
nx.draw_networkx(x, pos=pos, with_labels=False, node_color=node_color, node_size=node_size)
plt.axis('off')
```
